```python
from datetime import date
from datetime import datetime
```

```python
#Transactions

class Transaction:
    type = None          #debit or credit
    amount = None        #amount of transaction
    comments = None      #additional comments
    date = None          #date of transaction
    acc = None           #accont number 
    bank = None          #bank

    def addTransaction(myself,type,comments,amount,acc,bank,date=None):
        myself.type = type
        myself.amount = amount
        myself.comments = comments
        myself.acc = acc
        myself.bank = bank
        if date!=None:
          myself.date = date
        else:
          myself.date = date.today().strftime("%d/%m/%Y")
    

    def printTransaction(self):
        print(">>>>>>>>>>>>>><<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<")
        print("Type: "+self.type)
        print("amount: "+str(self.amount))
        print("comments: "+self.comments)
        print("date: "+self.date)
        print("acc: "+self.acc)
        print("bank: "+self.bank)
        print(">>>>>>>>>>>>>><<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<\n\n\n")



```

```python
#Month
class Month:
    month = None
    year = None
    allTransactions = []
    totalTransactions = None
    
    def addTransactionToMonth(self,transaction):
        self.allTransactions.append(transaction)
        #sorting the list of transaction according to date
        self.allTransactions.sort(key=lambda x: datetime.strptime(x.date, "%d/%m/%Y"), reverse=False)
    def printAllMonthTransactions(self):
        for i in self.allTransactions:
            print("\n------------------")
            i.printTransaction()
            print("------------------\n")
            
    
    
    

```
